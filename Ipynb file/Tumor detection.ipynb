```python
def splitdata(p , split):
    #csplitting p folder
     if not os.path.exists(p):
       os.mkdir(p)
       for dir in os.listdir(ROOT_DIR):
           os.makedirs(os.path.join(p, dir))
           for img in np.random.choice(a =os.listdir(os.path.join(ROOT_DIR,dir) ),
                                  size = (math.floor(split*number_of_images[dir])-5),
                                  replace = False):
            O = os.path.join(ROOT_DIR,dir,img)
            D = os.path.join(p,dir)

            shutil.copy(O,D)
            os.remove(O)


     else:
       #shutil.rmtree(p)
       print("folder already exists")
```

```python
import os
import cv2
import numpy as np
from keras.applications.mobilenet import preprocess_input

def preprocessingImages1(path):
    """
    input = path
    output  : preprocessed images
    """
    images = []
    labels = []
    target_size = (224, 224)

    # Iterate through each image in the directory
    for root, dirs, files in os.walk(path):
        for file in files:
            if file.endswith('.jpg') or file.endswith('.png') or file.endswith('.jpeg') or file.endswith('.tif'):  # Assuming images are JPEG or PNG format
                image_path = os.path.join(root, file)
                label = 1 if 'tumor' in root.lower() else 0  # Assuming tumor images are in a subdirectory containing 'tumor'

                # Open and preprocess the image using OpenCV
                image = cv2.imread(image_path)
                image = cv2.resize(image, target_size)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB (OpenCV reads in BGR)
                image = preprocess_input(image)

                images.append(image)
                labels.append(label)

    # Convert lists to NumPy arrays
    images = np.array(images)
    labels = np.array(labels)

    return images, labels

# Usage
images, labels = preprocessingImages1("/content/train")

# Print number of images
print("Number of images:", len(images))
count = 0;
for i in labels:
  if i == 1:
    count = count +1;

print(count);
```

### predicting from the model



```python
from keras.layers import Flatten,Dense
from keras.models import  Model,load_model
from keras.applications.mobilenet import MobileNet,preprocess_input
```

```python
base_model = MobileNet(input_shape = (224,224,3),include_top = False)
```

```python
for layer in base_model.layers:
   layer.trainable = False;
```

```python
X = Flatten()(base_model.output)
X = Dense(units =1,activation = 'sigmoid')(X)
model = Model(base_model.input,X)
```

```python
model.summary()
```

```python
model.compile(optimizer = 'rmsprop',loss =keras.losses.binary_crossentropy,metrics = ['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint,EarlyStopping
mc = ModelCheckpoint(filepath = "bestmodel.keras",monitor = 'val_accuracy',verbose = 1,save_best_only = True)
es = EarlyStopping(monitor = 'val_accuracy',min_delta = 0.01,patience = 3,verbose = 1)
cb  = [mc,es]
```

```python
hist = model.fit(train_data[0],train_data[1],steps_per_epoch = 8,epochs = 30,validation_data=(val_data[0],val_data[1]),validation_steps=16,callbacks = cb)
```

```python
model = load_model("/content/bestmodel.keras")
acc = model.evaluate(test_data[0],test_data[1])[1]
print(f"the accuaracy of model is : {acc*100}%")

```
